### 1. Phân phối Nhị thức

- Có số lần thử nghiệm cố định n: Bạn biết trước số lần thử nghiệm hoặc phép đo. 
- Mỗi lần thử chỉ có hai kết quả (thành công hoặc thất bại): Ví dụ, trả lời đúng hoặc sai trong một bài kiểm tra.
- Xác suất thành công p không đổi trong mọi lần thử: Xác suất thành công giữa các lần thử là như nhau.
- Các lần thử nghiệm là độc lập với nhau: Kết quả của lần thử này không ảnh hưởng đến kết quả của lần thử khác.

Ví dụ: Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng

- Số lần thử nghiệm hữu hạn: Có 10 câu hỏi trong bài thi, n = 10.
- Mỗi lần thử chỉ có hai kết quả (thành công hoặc thất bại): Mỗi câu hỏi chỉ có một phương án đúng, thành công là chọn đúng (p = 1/4), thất bại là chọn sai (1 - p = 3/4).
- Xác suất thành công không đổi trong mọi lần thử: Xác suất trả lời đúng mỗi câu là 1/4, không đổi cho mỗi câu hỏi.
- Các lần thử nghiệm là độc lập với nhau: Việc trả lời mỗi câu hỏi không ảnh hưởng đến kết quả các câu hỏi khác.

### 2. Phân phối Poisson

- Sự kiện xảy ra độc lập với nhau: Không có sự phụ thuộc giữa các sự kiện.
- Tốc độ trung bình λ không đổi: Số lượng sự kiện trung bình trong một đơn vị thời gian (hoặc không gian) là không đổi.
- Khoảng thời gian hoặc không gian xác định: Phân tích trong khoảng thời gian hoặc không gian nhất định.
- Xác suất xảy ra nhiều sự kiện trong khoảng thời gian ngắn là thấp: Xác suất nhận được số lượng lớn sự kiện trong một khoảng thời gian rất ngắn là thấp.

Ví dụ: Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút.

- Sự kiện xảy ra độc lập: Cuộc gọi đến không phụ thuộc vào cuộc gọi trước đó.
- Tốc độ trung bình không đổi: Trung bình 3 cuộc gọi mỗi phút, ký hiệu là λ=3.
- Số lượng sự kiện trong khoảng thời gian ngắn là thấp: Xác suất nhận được số lượng lớn cuộc gọi trong một khoảng thời gian rất ngắn (ví dụ, nhiều hơn 3 cuộc gọi trong một phút) là thấp, và xác suất này giảm theo số lượng cuộc gọi tăng lên.
- Khoảng thời gian xác định: Phân tích trong khoảng thời gian cụ thể là một phút.

### 3. Phân phối Chuẩn

- Xét tập hợp của nhiều biến ngẫu nhiên độc lập và có cùng phân phối: Theo định lý giới hạn trung tâm, tổng (hoặc trung bình) của các biến ngẫu nhiên độc lập, khi n đủ lớn, sẽ hội tụ về một phân phối chuẩn.
- Các biến được đo lường từ các quá trình hoặc hiện tượng tự nhiên: Nhiều biến trong thực tế (chiều cao, cân nặng, điểm số, v.v.) có xu hướng tuân theo phân phối chuẩn.
 - Cần mô tả sự phân bố của biến xung quanh một giá trị trung bình μ và độ lệch chuẩn 𝜎: Phân phối chuẩn được mô tả bởi hai thông số này và đối xứng quanh giá trị trung bình.
