### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
POP_2 = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])

mean = np.mean(POP_2)                
std_sample = np.std(POP_2)  

confidence = 0.95
alpha = 1 - confidence
```

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

```python
sigma = 0.3
z_score = norm.ppf(1 - alpha / 2)  
margin_of_error_a = z_score * (sigma / np.sqrt(len(POP_2)))
ci_a = (mean - margin_of_error_a, mean + margin_of_error_a)

print("\nTrường hợp a: Biết σ = 0.3")
print(f"Z-score: {z_score:.3f}")
print(f"Khoảng tin cậy 95%: {ci_a[0]:.3f} đến {ci_a[1]:.3f}")
```

b, Không biết $\sigma$

```python
t_score = t.ppf(1 - alpha / 2, df= len(POP_2) - 1)  
margin_of_error_b = t_score * (std_sample / np.sqrt(len(POP_2)))
ci_b = (mean - margin_of_error_b, mean + margin_of_error_b)

print("\nTrường hợp b: Không biết σ")
print(f"T-score: {t_score:.3f}")
print(f"Khoảng tin cậy 95%: {ci_b[0]:.3f} đến {ci_b[1]:.3f}")
```

Cho biết khoảng ước lượng bạn xây dựng có đúng không, dựa trên giá trị đã tính ở câu trên.

```python
df = pd.read_csv("data/SAT-lard.csv")

mean = df['SAT Score'].mean()
std_sample = df['SAT Score'].std()
n = len(df)

df.head()
```

a. Giả sử bộ dữ liệu là kết quả của quá trình lấy mẫu ngẫu nhiên từ các học sinh. Bạn hãy xây dựng khoảng tin cậy cho trung bình điểm thi của tất cả các sinh viên với độ tin cậy 99% (`Đáp án: 105.2 ± 3.10`)

```python
alpha = 1 - 0.99  
df_degrees = n - 1  
t_score = t.ppf(1 - alpha / 2, df=df_degrees)

margin_of_error = t_score * (std_sample / np.sqrt(n))
ci = (mean - margin_of_error, mean + margin_of_error)

print(f"Khoảng tin cậy 99%: {ci[0]:.3f} đến {ci[1]:.3f}")
print(f"Đáp án: {mean:.1f} ± {margin_of_error:.2f}")
```
